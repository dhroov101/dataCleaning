### Importing necessay libraries

```python
import pandas as pd
import numpy as np
```

### Reading the Data 

```python
df= pd.read_csv('C:/Users/ASUS/Downloads/data cleaning/Uncleaned_DS_jobs.csv')
df.head()
```

### Droping Unwanted Columns 
    >'Index' and 'Competitors' columns are unwanted data attributes
    Hence we drop both the columns

```python
df.drop('index',inplace=True , axis=1)
df.drop('Competitors',inplace=True , axis=1)
df.head()
```

### Cleaning and Converting Estimated Salary into integer

```python
df['Salary Estimate'].unique()
```

#### As it is visible here the salary ranges have unclean and garbage values


```python
df['Salary Estimate'].apply(lambda x: x.split('(')[0]).unique()
```

#### removing the "(Glassdoor est.)" garbage value

```python
df['Salary Estimate']=df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
df['Salary Estimate'].apply(lambda x: x.replace('$','').replace('K','')).unique()
```

#### Removing $ and K 

```python
df['Salary Estimate']=df['Salary Estimate'].apply(lambda x: x.replace('$','').replace('K',''))
```

#### Creating min and max salary columns


```python
df['min_salary']=df['Salary Estimate'].apply(lambda x: x.split('-')[0])
df['max_salary']=df['Salary Estimate'].apply(lambda x: x.split('-')[1])
```

```python
df.head()
```

#### Cleaning the Company name

    > The company names contain intergers which need to be removes

```python
df['Company Name']=df['Company Name'].apply(lambda x: x.split('\n')[0])
df.head()
```

#### Sorting on the basis of ratings and min salary

```python
df.sort_values(by=['Rating','min_salary'],ascending=[False,True])
```

#### Replacing negative ratings with 0

```python
df[df['Rating']==-1.0].shape
```

```python
df.Rating.dtype
```

```python
df.Rating = np.where(df.Rating==-1.0,0,df.Rating)
```

```python
df=df.sort_values(by=['Rating','min_salary'],ascending=[False,True])
```

#### Making a new column with the job state

```python
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[-1])
df.head()
```

#### New column with company age

```python
df['company_age'] = df.Founded.apply(lambda x: x if x<1 else 2023-x)
df.head()
```

```python
df[df['company_age']==-1].shape
```

#### Changing comapny age from negative to 0

```python
df.company_age = np.where(df.company_age==-1.0,0,df.Rating)
df[df['company_age']==-1].shape
```

```python
df.head()
```

```python
df['Job Description'][2].split('\n')
```

#### Creating attributes according to job description
    > if a particular skill is present in job description ,it is updated to 1
        else 0

```python
df['python'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df['hadoop'] = df['Job Description'].apply(lambda x: 1 if 'hadoop' in x.lower() else 0)
df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df['tableau'] = df['Job Description'].apply(lambda x: 1 if 'tableau' in x.lower() else 0)
df['big_data'] = df['Job Description'].apply(lambda x: 1 if 'big data' in x.lower() else 0)
df.head()
```

#### Simplifying Role name

```python
df['Job Title'].value_counts()
```

```python
def title_simplifier(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'mle'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'na'
    
```

```python
df['job_simp'] = df['Job Title'].apply(title_simplifier)

df.job_simp.value_counts()
```

#### Defining seniority

```python
def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'sr' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
            return 'senior'
    elif 'jr' in title.lower() or 'jr.' in title.lower():
        return 'jr'
    else:
        return 'na'
```

```python
df['seniority'] = df['Job Title'].apply(seniority)
df.seniority.value_counts()
```
